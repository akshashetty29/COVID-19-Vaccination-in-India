# COVID-19 Vaccination in India


```python
#Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#import dataset
df=pd.read_csv("datasets\covid_vaccine_india.csv",parse_dates=["Date"],dayfirst=True)
df.head()
```

```python
#Data Preprocessing
del df["AEFI"]
df=df.fillna(0)
df.head()
```

```python
#Totat dose administered in all states as per 18th May 2021
cur_date=df[df.Date=='2021-05-18']
max_total_doses=cur_date.sort_values(by="Total Doses Administered",ascending=False)
max_total_doses
```

```python
sns.set(rc={'figure.figsize':(10,15)})
sns.barplot(y="State",x="Total Doses Administered",data=max_total_doses)
plt.show()
```

```python
#Vaccination in progress Gender wise
total_male=df["Male(Individuals Vaccinated)"].sum()
total_female=df["Female(Individuals Vaccinated)"].sum()
total_trans=df["Transgender(Individuals Vaccinated)"].sum()
```

```python
sizes=[total_male,total_female,total_trans]
label=["Male","Female","Transgender"]
fig = plt.figure(figsize =(10, 7))
plt.pie(sizes, labels = label)
plt.legend()
plt.show()
```

```python
#Vaccination in progress Age wise
total_18=df["18-30 years (Age)"].sum()
total_30=df["30-45 years (Age)"].sum()
total_45=df["45-60 years (Age)"].sum()
total_60=df["60+ years (Age)"].sum()
```

```python
size1=[total_18,total_30,total_45,total_60]
label1=["18-30","30-45","45-60","60+"]
fig2 = plt.figure(figsize =(10, 7))
plt.pie(size1, labels = label1)
plt.legend()
plt.show()
```

```python
#Visulaising total vaccination that took place in maharashtra
maha=df[df.State=='Maharashtra']
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="Date",y="Total Doses Administered",data=maha,color="g")
plt.show()
```

```python
#Converting date to Ordinal Format
import datetime as dt
maha['Date']=maha['Date'].map(dt.datetime.toordinal)
maha.head()
```

```python
x=maha['Date'].values
y=maha['Total Doses Administered'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
#Training the model
lr.fit(x_train.reshape(-1,1),y_train.reshape(-1,1))
```

```python
n_pred=lr.predict(x_test.reshape(-1,1))
logreg_accuracy=lr.score((x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
print(logreg_accuracy)
```
